```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import yt
```

```python
num_samples = 30

rad, dens, eints, press, temps, times = [[] for _ in range(6)]
```

```python
frate = 6
```

```python
for i in range(num_samples):
    ds = yt.load(f"exports/ws2_hdf5_plt_cnt_{i:04d}")

    ad = ds.all_data()

    t = float(ds.current_time.to("s"))
    r = ad[("index","r")]*1e4 # to get micron
    rho = ad[("flash","dens")]  # in g/cm^3
    eint = ad[("flash","eint")]
    pres = ad[("flash","pres")]
    temp = ad[("flash","temp")]

    rad.append(r)
    dens.append(rho)
    eints.append(eint)
    press.append(pres)
    temps.append(temp)
    times.append(t)
```

```python
fig, ax = plt.subplots(figsize=(6,5), dpi=300)

def update(i):
    ax.clear()
    ax.set_xlabel(r"Radius ($\mu$m)", fontsize=16)
    ax.set_ylabel(r"Density (g/cm$^3$)", fontsize=16)
    ax.set_ylim(1e-4, 4)
    ax.set_yscale('log')
    ax.grid()
    ax.plot(rad[i], dens[i], label=f"t={times[i]:.2e} s")
    ax.legend(loc="upper left")
    ax.set_title('Density Profile from 1-D FLASH Simulation', fontsize=18)

plt.rcParams['animation.ffmpeg_path'] = '/opt/homebrew/bin/ffmpeg'
anim = animation.FuncAnimation(fig, update, frames=num_samples, interval=50)
FFwriter = animation.FFMpegWriter(fps=frate)
anim.save(f'density.mp4', writer=FFwriter)
plt.close()
```
